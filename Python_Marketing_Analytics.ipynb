### b. Understanding how many new customers acquired every month

```python
ORDERS['year_month_acquired'] = ORDERS.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x , format = "%Y-%m"))
ORDERS.head()
```

```python
# Merging the orders and customers so as to get the unique customers for each month for that year
customer_orders = pd.merge(CUSTOMERS , ORDERS , how = 'left' , on= 'customer_id')
customer_orders.head()
```

```python
customers_per_month = customer_orders.loc[:, ['customer_unique_id', 'year_month_acquired']] 
customers_per_month.drop_duplicates(subset='customer_unique_id' , inplace= True)
customers_per_month.head()
```

```python
new_customers_per_month = customers_per_month.groupby('year_month_acquired')[['customer_unique_id']].count()
display(new_customers_per_month.head())

# Plotting the unique customers acquired per month year-wise
new_customers_per_month.plot(kind = 'bar' , figsize =(8,5))
plt.title("New Customers Acquired Per Month Year-Wise")
plt.show()
```

### c)  Understand the retention of customers on month on month basis


```python
# We have the new customers per month count
new_customers_per_month.head()
```

```python
# Total customers monthly (from the merged orders and customer data)
customers_monthly = customer_orders.groupby('year_month_acquired')[['customer_unique_id']].count()
customers_monthly.head()
```

```python
# Merging the new and total monthly customers
Total_New_cust_monthly = pd.merge(new_customers_per_month , customers_monthly , how= 'left' , on ="year_month_acquired")
Total_New_cust_monthly.head()
```

```python
# Subtracting the total customers count from the new customers count 
Total_New_cust_monthly['Number_Retained_Customers'] = Total_New_cust_monthly.customer_unique_id_y - Total_New_cust_monthly.customer_unique_id_x
Total_New_cust_monthly.head()
```

### 2) Performing Customers/sellers Segmentation
### a) Divide the customers into groups based on the revenue generated 

```python
# Merging the customer orders and order items
customer_order_item = pd.merge(customer_orders , ORDER_ITEMS , how ='inner', on= 'order_id' )
# Computing the total price
customer_order_item['Total_Price'] = customer_order_item.price + customer_order_item.freight_value
display(customer_order_item.head())
```

```python
# revenue generated by the customers
revenue_by_cust = customer_order_item.groupby('customer_unique_id')['Total_Price'].sum().reset_index()
revenue_by_cust.rename(columns ={'Total_Price':'Total_Price'} , inplace = True)
revenue_by_cust.head()
```

```python
# Grouping the customers based on the revenue
revenue_by_cust['Customer_group_based_revenue'] = pd.cut(revenue_by_cust.Total_Price,
                                             bins= np.arange(0 , 16000, 5000),
                                             labels=['0-5000','5000-10000','10000-15000',])

display(revenue_by_cust)
```

### b) Divide the sellers into groups based on the revenue generated 

```python
# Getting the sellerid's and their prices
revenue_by_sellers = ORDER_ITEMS.groupby('seller_id')['price'].sum().reset_index()

# Renaming the columns for better understanding
revenue_by_sellers.rename( columns = {'price':'Total_Sold_Amount'}, inplace= True)
display(revenue_by_sellers.head())
```

```python
# Grouping the sellers based on revenue
revenue_by_sellers['sellers_group_based_revenue'] = pd.cut(revenue_by_sellers.Total_Sold_Amount,
                                                           bins= np.arange(0 , 310000, 100000),
                                                           labels=['0-100000','10000-200000','200000-300000'])

display(revenue_by_sellers.head())
```

### 3) Cross-Selling (Which products are selling together)

```python
# Getting the ordered items
ordered_items = pd.merge(ORDERS , ORDER_ITEMS , how='left' , on='order_id' )
ordered_items.head()
```

```python
# Cross selling , means different products that were bought at the same purchase or time and sort it accordingly

same_purchase_products = ordered_items[ordered_items.duplicated(['order_purchase_timestamp','customer_id'],keep=False)].sort_values(by='order_purchase_timestamp')

# Getting only the required columns
same_purchase_products = same_purchase_products.loc[:,['customer_id','order_purchase_timestamp','product_id' ]].reset_index()


# Removing the unwanted column index with random index numbers
same_purchase_products.drop(columns = 'index',axis=1, inplace=True)

# Merging with the products data to get the product category name 
same_purchase_product_name = pd.merge(same_purchase_products , PRODUCTS , how='left' , on='product_id' )

same_purchase_product_name = same_purchase_product_name.loc[:,['customer_id','order_purchase_timestamp','product_id', 'product_category_name' ]]

print("The cross selling products, i.e the products that are selling together are :")
display(same_purchase_product_name)
```

### 5. Customer satisfaction towards category & product
### a. Which categories (top 10) are maximum rated & minimum rated?


```python
# Getting the reviews of the orders
ordered_items_products_review = pd.merge(ordered_items_products, ORDER_REVIEW_RATINGS, how = 'left', on = 'order_id')

# Grouping by the product category and summing the review score 
prod_cat_review = ordered_items_products_review.groupby('product_category_name')['review_score'].sum().reset_index()

# Deleting the mixed up index column and getting the top 10
top_10_prod_cat_review = prod_cat_review.sort_values('review_score', ascending = False).reset_index().drop(columns = 'index').head(10)
print("The Top 10 product categories are :")
display(top_10_prod_cat_review)


# Deleting the mixed up index column and below 10
below_10_prod_cat_review = prod_cat_review.sort_values('review_score', ascending = True).reset_index().drop(columns = 'index').head(10)
print("The Below 10 product categories are :")
display(below_10_prod_cat_review)
```

### b) Which products (top10) are maximum rated & minimum rated?

```python
# Getting the products and their reviews
product_review = ordered_items_products_review.groupby('product_id')['review_score'].sum().reset_index()

# Sorting in the descending order
product_review = product_review.sort_values(by='review_score',ascending = False)

# Getting the top 10 products based on the review
top_10_product_review =  product_review.reset_index().drop(columns = 'index').head(10)
print("The top 10 rated products are:")
display(top_10_product_review)


# Sorting in the ascending order
product_review = product_review.sort_values(by='review_score',ascending = True)

# Getting the below 10 products based on the review
below_10_product_review =  product_review.reset_index().drop(columns = 'index').head(10)
print("The below 10 rated products are:")
display(below_10_product_review)
```

### c) Average rating by location, seller, product, category, month etc.


```python
# Merging orders, order items, customers, and ratings
cust_order_items_review = pd.merge(ORDERS, ORDER_REVIEW_RATINGS, on='order_id' , how='left')
cust_order_items_review = pd.merge(cust_order_items_review, CUSTOMERS, on='customer_id' , how='left')
cust_order_items_review = pd.merge(cust_order_items_review, ORDER_ITEMS, on='order_id' , how='left')

cust_order_items_review.head()
```

```python
# Average rating by LOCATION
location_avg_rating = cust_order_items_review.groupby('customer_state')['review_score'].mean().reset_index()
location_avg_rating.rename(columns={'customer_state':'Location', 'review_score':'Average_Rating'}, inplace =True)
print('Location wise average rating :')
display(location_avg_rating)
```

```python
# Average rating by SELLER
seller_avg_rating = cust_order_items_review.groupby('seller_id')['review_score'].mean().reset_index()
seller_avg_rating.rename(columns={'review_score':'Average_Rating'}, inplace =True)
print('Seller wise average rating :')
display(seller_avg_rating)
```

```python
# Average rating by MONTH

# Getting the months 
cust_order_items_review['month'] = cust_order_items_review.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x , format = "%B"))

month_avg_rating = cust_order_items_review.groupby('month')['review_score'].mean().reset_index()
month_avg_rating.rename(columns={'review_score':'Average_Rating'}, inplace =True)

print('Month wise average rating :')
display(month_avg_rating)


```

```python
# Average rating by PRODUCT

product_avg_rating = cust_order_items_review.groupby('product_id')['review_score'].mean().reset_index()
product_avg_rating.rename(columns={'review_score':'Average_Rating'}, inplace =True)
print('Product wise average rating :')
display(product_avg_rating)
```

```python
# Average rating by CATEGORY

# Grouping by the product category and summing the review score 
category_avg_rating = ordered_items_products_review.groupby('product_category_name')['review_score'].mean().reset_index()
category_avg_rating.rename(columns={'review_score':'Average_Rating'}, inplace =True)
print('Category wise average rating :')
display(category_avg_rating)

```
